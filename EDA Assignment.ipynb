```python
# Loading and Understanding the Data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset from Seaborn
df = pd.read_csv("Chocolate Sales.csv")
df
```

```python
# histogram for product
sns.histplot(df['Product'],bins=30)
plt.title("Total Products")
plt.show()
```

```python
# bar plot for Sales Person
sns.countplot(x='Sales Person',data=df, kde = True)
plt.title("Sales person")
plt.xticks(rotation=100)
plt.show()
```

```python
# countplot for countries
data=df.copy()
sns.countplot(x='Country',data=data)
plt.title("Countries")
plt.xticks([0,1],['Europe','Asia'])
plt.show()
```

```python
# Boxplot of age by survival
sns.boxplot(x='Product',y='Country',data=df)
plt.title("Product in countries")
plt.show()
```

```python
# select any numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['number'])

#compute correlation matrix
corr_matrix = numeric_df.corr()

# Heatmap of correlation
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Feature correlation Heatmap")
plt.show()
```
